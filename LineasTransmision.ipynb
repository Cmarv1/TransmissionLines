## 1.Corriente y Voltaje.

Para entender la propagación de ondas en una líneas de transmisión se trabaja con el modelo de **parámetros distribuidos** con el que se modela la línea con sus elementos circuitales R, L, G, C y se obtienen ecuaciones para el voltaje y la corriente que se propagan por la línea.

Las ondas de voltaje y corriente que salen de la fuente, pueden reflejarse en la carga:
- Onda incidente, viaja de la fuente a la carga.
- Onda reflejada, viaja de la carga a la fuente.

Las ecuaciones del voltaje y la corriente:

<br>
$$\large v(z,t) = Re \, \{ \hat V(z)e^{j \omega t} \} [V]$$
$$\large i(z,t) = Re \, \{ \hat I(z)e^{j \omega t} \} [A]$$
<br>

Las ecuaciones de los fasores del voltaje $\hat V(z)$ y la corriente $\hat I(z)$ son:

<br>
$$\large \hat V(z) = V_0^+ \, e^{-\gamma z} \, + V_0^- \, e^{\gamma z} \, [V]$$
$$\large \hat I(z) = I_0^+ \, e^{-\gamma z} \, + I_0^- \, e^{\gamma z} \, [A]$$<br>

Incluyendo las ecuaciones de los fasores obtenemos:

$$\large v(z,t) = V_0^+ \, e^{-\alpha z}cos(\omega t - \beta z) + V_0^- \, e^{\alpha z}cos(\omega t + \beta z) [V]$$

$$\large i(z,t) = I_0^+ \, e^{-\alpha z}cos(\omega t - \beta z) + I_0^- \, e^{\alpha z}cos(\omega t + \beta z) [A]$$<br>



donde:
- $z$ es la variable que representa el espacio.
- $V_0^+$, $I_0^+$ son las **amplitudes** de las ondas incidentes de voltaje y corriente respectivamente. <br>
- $V_0^-$, $I_0^-$ son las **amplitudes** de las ondas reflejadas de voltaje y corriente respectivamente. <br>
- $\gamma$ es la **constante de propagación**. <br>

## 2.Constante de propagación.

Al igual que se describió para la propagación del Campo Eléctrico en un material, $\gamma$, la constante de propagación se describe como: <br>

<br>
$$\large \gamma = \alpha + j \beta \, [1/m]$$<br>

donde
- $\alpha$ es la constante de atenuación [Np/m]. 
- $\beta$ es la constante de fase de la onda [rad/m].

### 2.1. Constante de atenuación

La constante de atenuación puede describirse en unidades lineales o en dB, la forma de relacionarlas es la siguiente:

<br>
$$\large \alpha [dB/z] = 8.68 \alpha [1/z]$$<br>

donde $z$ es la unidad espacial, que puede ser en metros [m] o kilómetros [km].
<br>

### 2.2. Constante de fase

Al propagarse la onda a través de un dieléctrico, la constante de fase puede calcularse con las ecuaciones de propagación en un dieléctrico. En este caso puede calcularse a partir de la siguiente ecuación para la velocidad de fase de la onda electromagnética:

<br>
$$\large v = \frac{\omega}{\beta} = \frac{c}{\sqrt{\epsilon_r}}$$<br>


## 3.Impedancia, Impedancia característica y Coeficiente de reflexión.

### 3.1. Convención

Para una línea de transmisión de longitud $l$ en metros, por convención en esta sección:
- La fuente se ubica en $z = -l$.
- La carga se ubica en $z = 0$.
<br>

Representaremos $Z_L$ como la impedancia de la carga.
<br><br>

### 3.2. Impedancia

La impadancia es valor complejo que se nombra como $Z(z)$, que se mide en Ohms $[\Omega]$ depende de $z$ (ubicación en la línea) y representa la impedancia vista hacia la derecha de la línea de transmisión a partir del punto $z$.

De acuerdo a la convención:

<br>
$$\large Z(0) = Z_L \, [\Omega]$$<br>

**Por ejemplo**, una línea de transmisión de $10 m$ de lontitud, que se conecta a una carga de $50 \Omega$, tenemos lo siguiente:
- La fuente se ubica en $z = 10m$.
- La carga se ubica en $z = 0$.
- La carga $Z(0) = Z_L = 50 \Omega$.
<br><br>

### 3.3. Impedancia característica

La impadancia característica es valor complejo que se nombra como $Z_0$, que se mide en Ohms $[\Omega]$ independiente de $z$ y es un valor asociado a la línea de transmisión.
<br><br>

### 3.4. Coeficiente de reflexión

El Coeficiente de reflexión es valor complejo que se nombra como $\Gamma (z)$, es adimensional y depende de $z$ (ubicación en la línea). Representa la proporción de la onda de Voltaje que se refleja desde el punto $z$ hacia la fuente.

De acuerdo a la convención $\Gamma (-l)$ es la proporción de la onda de voltaje que se refleja en la carga y se dirige hacia la fuente.

Para calcular el coeficiente de reflexión pueden emplearse las ecuaciones:

$$\large \Gamma(z) = \frac{V_0^-}{V_0^+} e^{2 \gamma z} = \frac{Z(z) - Z_0}{Z(z) + Z_0}$$<br>

La magnitud del coeficiente de reflexión varía entre $0$ y $1$:
$$ 0 \leq |\Gamma(z)| \leq 1$$
<br>

### 3.5. Ejemplo

Calcular el coeficiente de reflexión en la carga y la constante de fase de una línea de transmisión con las siguientes características:
- Longitud $l = 10m$.
- Impedancia de la carga $Z(0) = Z_L = 50 \Omega$.
- Impedancia característica $Z_0 = 75\Omega$.
- Amplitud de voltaje incidente $V_0^+=10 V$.
- Constante de atenuación $\alpha = 0.002 \, [1/m]$
- La onda tiene una frecuencia de $100 MHz$.
- La permitividad relativa del dieléctrico es $\epsilon_r = 2.2$.

Para los datos del ejemplo puede calcularse el coeficiente de reflexión en la carga $\Gamma(0)$:

<br>
$$\Gamma(0) = \frac{Z(0) - Z_0}{Z(0) + Z_0} = \frac{50 - 75}{50 + 75} = \frac{-25}{125} = -0.2$$<br>

¿Cómo se interpreta este valor?

Puede calcularse la amplitud de la onda de voltaje reflejada $V_0^-$, para $z=0$:

$$V_0^- = V_0^+ \cdot \Gamma(0) \cdot e^{0} = 10(-0.2)(1) = -2V$$

La constante de fase $\beta$, se calcula de la siguiente forma:

$$\beta = \frac{2\pi f \sqrt{\epsilon_r}}{v} = \frac{2\pi (1 \times 10^8 ) \sqrt{2.2}}{3 \times 10^8} = 3.11 \, [rad/m] $$

### 3.6. Ejemplo interactivo

Puede graficarse la onda incidente y la onda reflejada para el fasor del voltaje a lo largo de la línea de transmisión.

```
import matplotlib.pyplot as plt

Rcoef = 0.6  #@param {type: "slider", min: -1, max: 1, step: 0.1}
alpha = 0.1  #@param {type: "slider", min: 0, max: 0.1, step: 0.01}

f = 2e8
l = 10
Vomas = 1
Vomenos = Vomas*Rcoef
beta = 2*np.pi*f*np.sqrt(2.2)/3e8
omega = 2*np.pi*f
gamma = alpha + beta*1j

z = np.arange(-l,0,l/200)
Vi = Vomas*np.exp(-alpha*z)*np.cos(-beta*z)
Vr = Vomenos*np.exp(alpha*z)*np.cos(beta*z)

fig, ax = plt.subplots(2,1, figsize=(6, 10))
ax[0].plot(z, Vi, label='Incidente')
ax[0].plot(z, Vr, 'r--', label='Reflejada')
ax[0].axis([-10, 0, -2.5, 2.5])
ax[0].set_ylabel('v(z) [V]', fontsize=14)
ax[0].set_title('Ondas incidente y reflejada', fontsize=14)
ax[0].legend()

ax[1].plot(z, Vi + Vr)
ax[1].axis([-10, 0, -2.5, 2.5])
ax[1].set_xlabel('distancia z [m]', fontsize=14)
ax[1].set_ylabel('v(z) [V]', fontsize=14)
ax[1].set_title('Onda total', fontsize=14)

fig.show()
```

## 3.Potencia.

La Potencia es un valor real que se nombra como $P(z)$, se mide en Vatios [W] y depende de $z$ (ubicación en la línea). Representa la potencia medida en el punto $z$.

### 3.1. Convención

La convención para calcular la potencia es diferente a la que se presentó en la sección anterior:

- La fuente se ubica en $z = 0$.
- La carga se ubica en $z = l$.
<br>

### 3.2. Cálculo de la potencia

La ecuación para calcular la potencia en la ubicación $z$:

$$\large P(z) = P(0)e^{-2 \alpha z} \, [W]$$<br>

donde:
- $P(0)$ es la potencia a la entrada de la línea de transmisión, es decir, en $z=0$.
<br><br>

### 3.3. Ejemplo

Graficar la potencia a lo largo de una línea de transmisión con las siguientes características:
- Longitud $l = 10m$.
- Potencia a la entrada de la línea $P(0) = 10 W$.
- Constante de atenuación $\alpha = 0.02 \, [1/m]$.

```
#@title {display-mode: "form"}
l = 10
P0 = 10
alpha = 0.02

z = np.arange(0,l,l/100)
P = P0*np.exp(-2*alpha*z)

fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(z, P)
ax.axis([0, l, 0, 12])
ax.set_xlabel('distancia z [m]', fontsize=14)
ax.set_ylabel('P(z) [W]', fontsize=14)
ax.set_title('Potencia', fontsize=14)

fig.show()
```
